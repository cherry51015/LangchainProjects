```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ['OPENAI_API_KEY']=os.getenv('OPEN_API_KEY')
## Langsmith Tracking
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model='gpt-4o')
llm
```

```python
result=llm.invoke("What is generative AI?")
result
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt=ChatPromptTemplate.from_messages([('system':'You are an expert AI engineer provide me the answers based on the Professional high Level Experience'),('user':'{input}')])
prompt
```

```python
chain=prompt|llm
response=chain.invoke({"input":["can u tell me about Langsmith", "Why 90 percent of engineers fails to get Ai engineer roles"]})
response
```

```python
#when we want only string, there is output parser called string output parser
from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser()
chain=prompt|llm|output_parser

response=chain.invoke({"input":"Can you tell me about Langsmith?"})
print(response)
```
